```python
# Function to categorize diagnosis data
def categorize_diagnosis_data(df, column_name):
    diagnoses_data = []
    treatments_data = []
    pathology_details_data = []

    for entry in df[column_name]:
        if isinstance(entry, list):  # If data is stored as a list in the column
            for item in entry:
                diagnosis_id = item.get('diagnosis_id', None)

                # Diagnosis Data
                diagnoses_data.append({
                    'diagnosis_id': diagnosis_id,
                    'primary_diagnosis': item.get('primary_diagnosis', None),
                    'age_at_diagnosis': item.get('age_at_diagnosis', None),
                    'ajcc_pathologic_stage': item.get('ajcc_pathologic_stage', None),
                    'ajcc_pathologic_t': item.get('ajcc_pathologic_t', None),
                    'ajcc_pathologic_n': item.get('ajcc_pathologic_n', None),
                    'ajcc_pathologic_m': item.get('ajcc_pathologic_m', None),
                    'method_of_diagnosis': item.get('method_of_diagnosis', None),
                    'metastasis_at_diagnosis': item.get('metastasis_at_diagnosis', None),
                    'site_of_resection_or_biopsy': item.get('site_of_resection_or_biopsy', None),
                    'icd_10_code': item.get('icd_10_code', None),
                    'state': item.get('state', None)
                })

                # Treatments Data
                if 'treatments' in item:
                    for treatment in item['treatments']:
                        treatments_data.append({
                            'diagnosis_id': diagnosis_id,
                            'treatment_id': treatment.get('treatment_id', None),
                            'treatment_type': treatment.get('treatment_type', None),
                            'treatment_intent_type': treatment.get('treatment_intent_type', None),
                            'treatment_or_therapy': treatment.get('treatment_or_therapy', None),
                            'treatment_dose': treatment.get('treatment_dose', None),
                            'prescribed_dose_units': treatment.get('prescribed_dose_units', None),
                            'days_to_treatment_start': treatment.get('days_to_treatment_start', None),
                            'route_of_administration': treatment.get('route_of_administration', None),
                            'therapeutic_agents': treatment.get('therapeutic_agents', None),
                            'treatment_outcome': treatment.get('treatment_outcome', None),
                            'state': treatment.get('state', None)
                        })

                # Pathology Details Data
                if 'pathology_details' in item:
                    for pathology in item['pathology_details']:
                        pathology_details_data.append({
                            'diagnosis_id': diagnosis_id,
                            'pathology_detail_id': pathology.get('pathology_detail_id', None),
                            'lymph_nodes_tested': pathology.get('lymph_nodes_tested', None),
                            'lymph_nodes_positive': pathology.get('lymph_nodes_positive', None),
                            'consistent_pathology_review': pathology.get('consistent_pathology_review', None),
                            'state': pathology.get('state', None)
                        })

    # Creating separate DataFrames
    df_diagnoses = pd.DataFrame(diagnoses_data)
    df_treatments = pd.DataFrame(treatments_data)
    df_pathology_details = pd.DataFrame(pathology_details_data)

    return df_diagnoses, df_treatments, df_pathology_details
```

```python
# Apply function to categorize data
df_diagnoses, df_treatments, df_pathology_details = categorize_diagnosis_data(df, 'diagnoses')

# Display the categorized DataFrames
print("Diagnosis Data:")
df_diagnoses.head(3)
```

### C50 Codes (Breast Cancer)
Each icd_10_code represents a specific type or location of breast cancer:

| ICD-10 Code | Description                                           |
|--------------|-------------------------------------------------------|
| C50.9        | Malignant neoplasm of breast, unspecified (General breast cancer) |
| C50.2        | Malignant neoplasm of upper-inner quadrant of breast |
| C50.3        | Malignant neoplasm of lower-inner quadrant of breast |
| C50.4        | Malignant neoplasm of upper-outer quadrant of breast |
| C50.5        | Malignant neoplasm of lower-outer quadrant of breast |
| C50.8        | Malignant neoplasm of overlapping sites of breast   |
| C50.919      | Malignant neoplasm of unspecified female breast      |      |


**Categorization Strategy**

| Category                         | ICD-10 Codes Included  | Description                              |
|----------------------------------|-------------------------|------------------------------------------|
| General Breast Cancer (Unspecified) | C50.9, C50.919          | Cancer of unspecified breast region      |
| Upper Breast Quadrant Cancer     | C50.2, C50.4            | Cancer in the upper-inner or upper-outer quadrant |
| Lower Breast Quadrant Cancer     | C50.3, C50.5            | Cancer in the lower-inner or lower-outer quadrant |
| Overlapping or Multiple Quadrants | C50.8                   | Cancer involving multiple breast regins |


```python
# Define a mapping dictionary
icd_10_mapping = {
    'C50.9': 'General Breast Cancer',
    'C50.919': 'General Breast Cancer',

    'C50.2': 'Upper Breast Quadrant Cancer',
    'C50.4': 'Upper Breast Quadrant Cancer',

    'C50.3': 'Lower Breast Quadrant Cancer',
    'C50.5': 'Lower Breast Quadrant Cancer',

    'C50.8': 'Overlapping or Multiple Quadrants'
}

# Apply the mapping to the icd_10_code column
df_new['icd_category'] = df_new['icd_10_code'].map(icd_10_mapping)

print(df_new['icd_category'].value_counts())
print(df_new['icd_category'].nunique())
```
